```python
import gzip

def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield eval(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')

df = getDF('meta_Clothing_Shoes_and_Jewelry.json.gz')
```

```python
from scipy.stats import geom
```

```python
def truncated_geom_rvs(p, n):
    k = geom.rvs(p)
    if k <= n:
        return k
    else:
        return n
```

```python
d = {}
for j in range(1,6):
    d[j] = 0

for j in range(10**5):
    k = truncated_geom_rvs(.5, 5)
    d[k] += 1
```
